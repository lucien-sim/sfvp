# Create fake coordinates, u, v

```python
x = np.linspace(0, 360, 181)
y = np.linspace(0, 180, 91)

X, Y = np.meshgrid(x, y)

u = (np.sin(X / 20) - np.sin(Y / 20)) * 10
v = (np.cos(X / 20 + 5) + np.sin(Y / 20 - 5)) * 10

dx = x[1] - x[0]
dy = y[1] - y[0]
```

# Create guess at chi (velocity potential), psi (stream function)

They are on the same grid but have two extra values in the X and Y directions. 

```python
chi0 = np.random.normal(0, 1, size=(X.shape[0] + 2, X.shape[1] + 2))
psi0 = np.random.normal(0, 1, size=(X.shape[0] + 2, X.shape[1] + 2))
```

# Set up optimization

```python
def combine_psi_chi(psi: np.ndarray, chi: np.ndarray) -> np.ndarray:
    return np.hstack([psi.ravel(), chi.ravel()])


def extract_psi_chi(params: np.ndarray, shape_psi: tuple) -> tuple: 
    psi = params[:np.product(shape_psi)].reshape(shape_psi)
    chi = params[np.product(shape_psi):].reshape(shape_psi)
    return psi, chi


def shift2d(x: np.ndarray, n: int, axis: int, fill_val: any = 0.) -> np.ndarray: 
    x = np.roll(x, shift=n, axis=axis)
    if n > 0 and axis == 0: 
        x[:n, :] = fill_val
    elif n < 0 and axis == 0: 
        x[n:, :] = fill_val
    elif n > 0 and axis == 1: 
        x[:, :n] = fill_val
    elif n < 0 and axis == 1: 
        x[:, n:] = fill_val
    return x


def calculate_velocity(psi: np.ndarray, chi: np.ndarray, dx: float, dy: float) -> tuple: 
    uhat = (chi[2:, 1:-1] - chi[:-2, 1:-1]) / (2 * dx) - (psi[1:-1, 2:] - psi[1:-1, :-2]) / (2 * dy)
    vhat = (chi[1:-1, 2:] - chi[1:-1, :-2]) / (2 * dy) + (psi[2:, 1:-1] - psi[:-2, 1:-1]) / (2 * dx)
    return uhat, vhat


def calculate_irrotational_flow(psi: np.ndarray, chi: np.ndarray, dx: float, dy: float) -> tuple: 
    u_irrot = (chi[2:, 1:-1] - chi[:-2, 1:-1]) / (2 * dx)
    v_irrot = (chi[1:-1, 2:] - chi[1:-1, :-2]) / (2 * dy)
    return u_irrot, v_irrot


def calculate_nondivergent_flow(psi: np.ndarray, chi: np.ndarray, dx: float, dy: float) -> tuple: 
    u_nondiv = -(psi[1:-1, 2:] - psi[1:-1, :-2]) / (2 * dy)  # SIGN HERE?
    v_nondiv = (psi[2:, 1:-1] - psi[:-2, 1:-1]) / (2 * dx)
    return u_nondiv, v_nondiv


def cost_fcn(params: np.ndarray, u: np.ndarray, v: np.ndarray, dx: float, dy: float, lam: float, shape_psi: tuple) -> float:
    psi, chi = extract_psi_chi(params, shape_psi)
    uhat, vhat = calculate_velocity(psi, chi, dx, dy)
    return (np.mean((uhat - u) ** 2 + (vhat - v) ** 2) + np.mean(lam * (psi ** 2 + chi ** 2)))


def jac(params: np.ndarray, u: np.ndarray, v: np.ndarray, dx: float, dy: float, lam: float, shape_psi: tuple) -> np.ndarray: 
    
    psi, chi = extract_psi_chi(params, shape_psi)
    uhat, vhat = calculate_velocity(psi, chi, dx, dy)
    
    # Calculate portions of jacobian that link psi/chi to u/v. 
    err_u = np.pad((uhat - u), pad_width=1, mode='constant')
    err_v = np.pad((vhat - v), pad_width=1, mode='constant')
    jac_chi = (-shift2d(err_u, n=-1, axis=0) + shift2d(err_u, n=1, axis=0)) / dx + \
              (-shift2d(err_v, n=-1, axis=1) + shift2d(err_v, n=1, axis=1)) / dy
    jac_psi = (-shift2d(err_v, n=-1, axis=0) + shift2d(err_v, n=1, axis=0)) / dx + \
              (shift2d(err_u, n=-1, axis=1) - shift2d(err_u, n=1, axis=1)) / dy
    jac_psi_chi = combine_psi_chi(jac_psi, jac_chi)
    
    # Add portion related to magnitude of psi/chi
    jac = jac_psi_chi + 2 * lam * params
    
    return jac
```

## Make sure everything works as expected

```python
init_params = combine_psi_chi(psi0, chi0)
lam = 1e-10
eta = 0.05
shape_psi = psi0.shape

cf = list()
new_params = np.copy(init_params)
cf.append(cost_fcn(new_params, u, v, dx, dy, lam, shape_psi))
for _ in range(10000): 
    new_params = new_params - eta * jac(new_params, u, v, dx, dy, lam, shape_psi)
    cf.append(cost_fcn(new_params, u, v, dx, dy, lam, shape_psi))
    
psif, chif = extract_psi_chi(new_params, shape_psi)
    
plt.plot(range(len(cf)), cf)
plt.xlabel('iterations')
plt.ylabel('cost function')
```

```python
from scipy import optimize

init_params = combine_psi_chi(psi0, chi0)
lam = 1e-6
shape_psi = psi0.shape

opt = optimize.minimize(
    fun=cost_fcn, 
    x0=init_params,
    args=(u, v, dx, dy, lam, shape_psi), 
    method='Newton-CG', 
    jac=jac
)

psif, chif = extract_psi_chi(opt.x, shape_psi)

print(opt)
```

# Rebuild on staggered grid

```python
from scipy import integrate


def calculate_velocity2(psi: np.ndarray, chi: np.ndarray, dx: float, dy: float) -> tuple: 
    uhat = ((chi[1:, :-1] - chi[:-1, :-1]) / dx + (chi[1:, 1:] - chi[:-1, 1:]) / dx -
            (psi[:-1, 1:] - psi[:-1, :-1]) / dy - (psi[1:, 1:] - psi[1:, :-1]) / dy) * 0.5
    vhat = ((chi[:-1, 1:] - chi[:-1, :-1]) / dy + (chi[1:, 1:] - chi[1:, :-1]) / dy + 
            (psi[1:, :-1] - psi[:-1, :-1]) / dx + (psi[1:, 1:] - psi[:-1, 1:]) / dx) * 0.5
    return uhat, vhat


def calculate_irrotational_flow2(psi: np.ndarray, chi: np.ndarray, dx: float, dy: float) -> tuple: 
    u_irrot = ((chi[1:, :-1] - chi[:-1, :-1]) / dx + (chi[1:, 1:] - chi[:-1, 1:]) / dx) * 0.5
    v_irrot = ((chi[:-1, 1:] - chi[:-1, :-1]) / dy + (chi[1:, 1:] - chi[1:, :-1]) / dy) * 0.5
    return u_irrot, v_irrot


def calculate_nondivergent_flow2(psi: np.ndarray, chi: np.ndarray, dx: float, dy: float) -> tuple: 
    u_nondiv = -((psi[:-1, 1:] - psi[:-1, :-1]) / dy + (psi[1:, 1:] - psi[1:, :-1]) / dy) * 0.5
    v_nondiv = ((psi[1:, :-1] - psi[:-1, :-1]) / dx + (psi[1:, 1:] - psi[:-1, 1:]) / dx) * 0.5
    return u_nondiv, v_nondiv


def first_guess(u, v, dx, dy): 
    
    # Guess psi (sf) once, using dpsi = -u_psi * dy, with boundary = dpsi = v_psi * dx
    intx = integrate.cumulative_trapezoid(v, dx=dx, axis=0, initial=0)[0]
    inty = -integrate.cumulative_trapezoid(u, dx=dy, axis=1, initial=0)
    psi1 = intx + inty
    
    # Guess psi(sf) again, using dpsi = v_psi * dx with boundary = dpsi = -u_psi * dy
    intx = integrate.cumulative_trapezoid(v, dx=dx, axis=0, initial=0)
    inty = -integrate.cumulative_trapezoid(u, dx=dy, axis=1, initial=0)[0:1, 0]
    psi2 = intx + inty
    
    # Average results and divide by 2
    psi_init = 0.5 * 0.5 * (psi1 + psi2)
    psi_init = psi_init - np.mean(psi_init)
    
    # Guess chi (vp) once, using dchi = v_chi * dy, with boundary = dchi = u_chi * dx
    intx = integrate.cumulative_trapezoid(u, dx=dx, axis=0, initial=0)[0]
    inty = integrate.cumulative_trapezoid(v, dx=dy, axis=1, initial=0)
    chi1 = intx + inty
    
    # Guess chi (vp) again, using dchi = u_chi * dx, with boundary = dchi = v_chi * dy
    intx = integrate.cumulative_trapezoid(u, dx=dx, axis=0, initial=0)
    inty = integrate.cumulative_trapezoid(v, dx=dy, axis=1, initial=0)[0:1, 0].ravel()
    chi2 = intx + inty
    
    # Average the results, then divide by 2 again
    chi_init = 0.5 * 0.5 * (chi1 + chi2)
    chi_init = chi_init - np.mean(chi_init)
    
    # Pad with zeros so shape is correct
    psi_init = np.pad(psi_init, pad_width=(0, 1), mode='constant')
    chi_init = np.pad(chi_init, pad_width=(0, 1), mode='constant')
    
    return psi_init, chi_init


def cost_fcn2(params: np.ndarray, u: np.ndarray, v: np.ndarray, dx: float, dy: float, lam: float, shape_psi: tuple) -> float:
    psi, chi = extract_psi_chi(params, shape_psi)
    uhat, vhat = calculate_velocity2(psi, chi, dx, dy)
    return (np.sum((uhat - u) ** 2 + (vhat - v) ** 2) + np.sum(lam * (psi ** 2 + chi ** 2)))


def jac2(params: np.ndarray, u: np.ndarray, v: np.ndarray, dx: float, dy: float, lam: float, shape_psi: tuple) -> np.ndarray: 
    
    psi, chi = extract_psi_chi(params, shape_psi)
    uhat, vhat = calculate_velocity2(psi, chi, dx, dy)
    
    # Calculate portions of jacobian that link psi/chi to u/v. 
    err_u = np.pad((uhat - u), pad_width=(0, 1), mode='constant')
    err_v = np.pad((vhat - v), pad_width=(0, 1), mode='constant')
    jac_chi = -(err_u - shift2d(err_u, n=1, axis=0)) / dx - shift2d((err_u - shift2d(err_u, n=1, axis=0)) / dx, n=1, axis=1) \
              -(err_v - shift2d(err_v, n=1, axis=1)) / dy - shift2d((err_v - shift2d(err_v, n=1, axis=1)) / dy, n=1, axis=0)
    jac_psi = -(err_v - shift2d(err_v, n=1, axis=0)) / dx - shift2d((err_v - shift2d(err_v, n=1, axis=0)) / dx, n=1, axis=1) \
              +(err_u - shift2d(err_u, n=1, axis=1)) / dy + shift2d((err_u - shift2d(err_u, n=1, axis=1)) / dy, n=1, axis=0)
    jac_psi_chi = combine_psi_chi(jac_psi, jac_chi)
    
    # Add portion related to magnitude of psi/chi
    jac = jac_psi_chi + 2 * lam * params
    
    return jac
```

```python
psi02, chi02 = first_guess(u, v, X, Y)

init_params = combine_psi_chi(psi02, chi02)
lam = 1e-10
eta = 0.015
shape_psi = psi02.shape

cf = list()
new_params = np.copy(init_params)
cf.append(cost_fcn2(new_params, u, v, dx, dy, lam, shape_psi))
for _ in range(5000): 
    new_params = new_params - eta * jac2(new_params, u, v, dx, dy, lam, shape_psi)
    cf.append(cost_fcn2(new_params, u, v, dx, dy, lam, shape_psi))
    
psif2, chif2 = extract_psi_chi(new_params, shape_psi)
    
plt.plot(range(len(cf)), cf)
plt.xlabel('iterations')
plt.ylabel('cost function')
```
